# Practica 12 - Crga de datos

# Cargar el CSV omitiendo la última fila

```python
import pandas as pd

ruta = "/content/drive/MyDrive/ClassFiles/miaad-nyc-r-s.csv"

df = pd.read_csv(ruta, skipfooter=1, engine='python')
```

# Verificacion de que NO se cargó la última línea

```python
df.tail(3)
```

# Cantidad de filas y colunmnas

```python
print("Filas:", df.shape[0])
print("Columnas:", df.shape[1])
```

# Practica 13 - Cambiar de nombre a la columna

# Impresion de nombres actuales

```python
print(df.columns)
```

# Aseguramiento de que la primera columna sea INDEX MIAAD

```python
if df.columns[0] != "INDEX MIAAD":
    df.rename(columns={df.columns[0]: "INDEX MIAAD"}, inplace=True)

print(df.columns)
```

# Practica 14 - Normalizacion de Nombres

```python
df.columns = df.columns.str.lower().str.replace(" ", "_")

print(df.columns)
```

# Practica 15 - Errores en los tipos de datos

```python
print(df.dtypes)
```

# Verificar tipo actual de columnas importantes

```python
print("sale_price:", df['sale_price'].dtype)
print("land_square_feet:", df['land_square_feet'].dtype)
print("gross_square_feet:", df['gross_square_feet'].dtype)
```

# Conversion a tipo flotantes

```python
df['sale_price'] = pd.to_numeric(df['sale_price'], errors='coerce')
df['land_square_feet'] = pd.to_numeric(df['land_square_feet'], errors='coerce')
df['gross_square_feet'] = pd.to_numeric(df['gross_square_feet'], errors='coerce')
```

# Verificacion

```python
print(df.dtypes)
```

# Practica 16 - Diccionario de datos

```python
diccionario_datos = {}

for columna in df.columns:
    diccionario_datos[columna] = str(df[columna].dtype)

print(diccionario_datos)
```
