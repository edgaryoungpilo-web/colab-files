# Practica 3 - Creacion de un arreglo y operaciones basicas

```python
import random

# 1. Crear un arreglo unidimensional de 20 elementos aleatorios entre 1 y 100
arreglo = [random.randint(1, 100) for _ in range(20)]

print("Arreglo original:")
print(arreglo)

# 2. Valor máximo
valor_maximo = max(arreglo)
print("\nValor máximo:", valor_maximo)

# 3. Valor mínimo
valor_minimo = min(arreglo)
print("Valor mínimo:", valor_minimo)

# 4. Media del arreglo
media = sum(arreglo) / len(arreglo)
print("Media del arreglo:", media)

# 5. Sumar 10 a cada elemento del arreglo
arreglo_modificado = [x + 10 for x in arreglo]

print("\nArreglo después de sumar 10 a cada elemento:")
print(arreglo_modificado)
```
