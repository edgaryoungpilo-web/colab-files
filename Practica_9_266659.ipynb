# Práctica 9: Distribución de Datos con plt.subplots()


# Paso 1 Importacion de librerias

```python
# Importar librerías
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

#Paso 2 Generador de datos aleatorios

```python
# Generar datos aleatorios
np.random.seed(42)  # Para reproducibilidad

data = {
    "Normal": np.random.normal(loc=0, scale=1, size=1000),
    "Uniforme": np.random.uniform(low=0, high=1, size=1000),
    "Exponencial": np.random.exponential(scale=1, size=1000),
    "Poisson": np.random.poisson(lam=3, size=1000)
}
```

# Paso 3 Creacion de DataFrame

```python
# Crear DataFrame
df = pd.DataFrame(data)
```

# Paso 4 - Figuras con subplots 2x2 y graficar histogramas en cada subplot (Distribuciones Normal,Uniforme, exponencialy possion)

```python
# Creacion de figura con subplots 2x2
fig, ax = plt.subplots(2, 2, figsize=(10, 8))

# Grafica de distribucion normal
ax[0, 0].hist(df["Normal"], bins=30)
ax[0, 0].set_title("Distribución Normal")

# Grafica de distribucion uniforme
ax[0, 1].hist(df["Uniforme"], bins=30)
ax[0, 1].set_title("Distribución Uniforme")

# Grafica de distribucion exponencial
ax[1, 0].hist(df["Exponencial"], bins=30)
ax[1, 0].set_title("Distribución Exponencial")

# Grafica de distribucion de Poisson
ax[1, 1].hist(df["Poisson"], bins=30)
ax[1, 1].set_title("Distribución Poisson")

# 5 Ajustar disposición
plt.tight_layout()

# Mostrar gráfica
plt.show()
```
