# Práctica 10: Análisis de Ventas con GridSpec

# Importacion de librerias

```python
# Paso 1 - Librerias
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec
```

# DataFrame

```python
# Paso 2 - Creacion de DataFrame de ventas
np.random.seed(42)

meses = ['Ene', 'Feb', 'Mar', 'Abr', 'May', 'Jun',
         'Jul', 'Ago', 'Sep', 'Oct', 'Nov', 'Dic']

ventas = np.random.randint(2000, 5000, 12)
gastos = np.random.randint(1000, 4000, 12)

df = pd.DataFrame({
    'Mes': meses,
    'Ventas': ventas,
    'Gastos': gastos
})
```

# Figura Gridspec c/ grafico de barras

```python
# Paso 3 - Crear figura con GridSpec
fig = plt.figure(figsize=(12, 8))
gs = GridSpec(2, 3, figure=fig)

# Gráfico (barra, ocupando dos columnas)
ax1 = fig.add_subplot(gs[0, 0:2])
ax1.bar(df['Mes'], df['Ventas'])
ax1.set_title('Ventas Mensuales')
ax1.set_ylabel('Ventas')
ax1.tick_params(axis='x', rotation=45)

# Ajustar espacios
plt.tight_layout()
plt.show()
```

# Figura Gridspec c/ grafico de linea

```python

# Paso 4 - Crear figura con GridSpec
fig = plt.figure(figsize=(12, 8))
gs = GridSpec(2, 3, figure=fig)

# Gráfico (línea, una columna)
ax2 = fig.add_subplot(gs[0, 2])
ax2.plot(df['Mes'], df['Gastos'], marker='o')
ax2.set_title('Gastos Mensuales')
ax2.set_ylabel('Gastos')
ax2.tick_params(axis='x', rotation=45)


# Ajustar espacios
plt.tight_layout()
plt.show()
```

# Figura Gridspec c/ grafico de dispersion

```python
# Paso 5 - Crear figura con GridSpec
fig = plt.figure(figsize=(12, 8))
gs = GridSpec(2, 3, figure=fig)

# Gráfico de dispersión, ocupando dos filas)
ax3 = fig.add_subplot(gs[:, 1:])
ax3.scatter(df['Ventas'], df['Gastos'])
ax3.set_title('Relación Ventas vs Gastos')
ax3.set_xlabel('Ventas')
ax3.set_ylabel('Gastos')

# Ajustar espacios
plt.tight_layout()
plt.show()
```
