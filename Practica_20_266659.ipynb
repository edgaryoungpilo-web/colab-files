# Punto 1. Introducción del Dataset

**Nombre del dataset:** Air Quality Dataset  

**Fuente:** Dataset público comúnmente utilizado para análisis ambiental  

**Descripción:**  
Este dataset contiene mediciones de calidad del aire, incluyendo contaminantes como CO, NOx, y variables ambientales como temperatura y humedad.

**Número de observaciones y variables:**  
Se determinará al cargar el dataset.

**Descripción de variables principales:**
- Date: Fecha de medición
- Time: Hora
- CO(GT): Monóxido de carbono
- PT08.S1(CO): Sensor de CO
- NMHC(GT): Hidrocarburos no metánicos
- C6H6(GT): Benceno
- NOx(GT): Óxidos de nitrógeno
- NO2(GT): Dióxido de nitrógeno
- T: Temperatura
- RH: Humedad relativa
- AH: Humedad absoluta

Carga de datos

```python
import pandas as pd

# Cargar dataset (sube el archivo a Colab primero)
df = pd.read_csv('/content/drive/MyDrive/ClassFiles/AirQuality.csv', sep=';')

# Mostrar primeras filas
df.head()
```

# Punto 2. Exploracion

```python
# Valores faltantes
df.isnull().sum()
```

```python
# Estadísticas descriptivas
df.describe()
```

Exploración inicial

Se observan valores faltantes en varias columnas, lo cual puede afectar el análisis.

Los tipos de datos incluyen variables numéricas y categóricas (fecha y hora).

Las estadísticas descriptivas muestran la dispersión de los datos y posibles valores extremos.

```python
# Reemplazar valores -200 (faltantes en este dataset)
df.replace(-200, pd.NA, inplace=True)

# Convertir a numérico
df = df.apply(pd.to_numeric, errors='ignore')
```

# Punto 3. Analisis univariado

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

Histograma

```python
plt.figure()
sns.histplot(df['CO(GT)'].dropna(), kde=True)
plt.title("Distribución de CO(GT)")
plt.show()
```

Boxplot

```python
plt.figure()
sns.boxplot(x=df['CO(GT)'])
plt.title("Boxplot de CO(GT)")
plt.show()
```

Análisis Univariado

La distribución de CO(GT) muestra concentración en valores bajos con algunos valores extremos.

El boxplot confirma la presencia de outliers.

# Punto 4. Analisis Bivariado

```python
plt.figure()
plt.scatter(df['T'], df['CO(GT)'])
plt.xlabel("Temperatura")
plt.ylabel("CO(GT)")
plt.title("Relación entre Temperatura y CO")
plt.show()
```

Análisis Bivariado

Se observa una ligera relación entre la temperatura y los niveles de CO, aunque no es completamente lineal.

# Punto 5. Analisis multivariado

```python
# Correlación
corr = df.corr()

plt.figure()
sns.heatmap(corr, annot=True)
plt.title("Mapa de calor - correlaciones")
plt.show()
```

Identificación de patrones

**Patrones encontrados:**
- Algunas variables presentan correlación positiva.
- CO y NOx parecen estar relacionados.

**Relaciones:**
- Variables de contaminación tienden a moverse juntas.

**Clusters:**
- No se observan clusters claros, pero sí agrupaciones en rangos similares.

**Outliers:**
- Existen valores extremos en varias variables.

**Variables relevantes:**
- CO(GT), NOx(GT), NO2(GT) son clave para analizar la calidad del aire.
