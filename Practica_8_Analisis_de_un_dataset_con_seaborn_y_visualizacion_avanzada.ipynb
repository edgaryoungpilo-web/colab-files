# Seccion 1 - Carga y exploracion de Database

```python
# Importacion de librerias
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Carga de database
df = sns.load_dataset('tips')

#Exploracion
df.head()
df.info()
df.describe()

# Valores nulos
df.isnull().sum()
```

# Visualizaciones de Seaborn


```python
# Mapa de calor
# Seleccion de variables numéricas
corr = df[['total_bill', 'tip', 'size']].corr()

plt.figure(figsize=(6,4))
sns.heatmap(
    corr,
    annot=True,
    cmap='coolwarm',
    linewidths=0.5,
    fmt=".2f"
)

plt.title('Mapa de Calor de Correlación')
plt.show()
```

```python
# Diagrama de Violin
plt.figure(figsize=(8,5))
sns.violinplot(
    x='day',
    y='tip',
    data=df,
    palette='Set2'
)

plt.title('Distribución de Propinas por Día')
plt.xlabel('Día de la semana')
plt.ylabel('Propina')
plt.show()
```

```python
# Grafica de dispersion
plt.figure(figsize=(8,5))
sns.scatterplot(
    x='total_bill',
    y='tip',
    hue='day',
    data=df,
    palette='tab10'
)

plt.title('Relación entre Total de la Cuenta y Propina')
plt.xlabel('Total de la cuenta')
plt.ylabel('Propina')
plt.legend(title='Día')
plt.show()
```

```python
# Grafica de barras
plt.figure(figsize=(6,4))
sns.barplot(
    x='sex',
    y='tip',
    data=df,
    palette='pastel',
    estimator='mean'
)

plt.title('Promedio de Propinas por Género')
plt.xlabel('Género')
plt.ylabel('Propina Promedio')
plt.show()
```
