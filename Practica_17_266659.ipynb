# Práctica 17: Limpieza y Transformación de Datos Categóricos

# Importacion de archivo

```python
import pandas as pd

ruta = "/content/drive/MyDrive/ClassFiles/nls97.csv"
df = pd.read_csv(ruta)

df.head()
```

# Ejercicio 1: Errores tipograficos y valores inconsistentes

#Valores unicos

```python
df['maritalstatus'].unique()
```

# Frecuencia de cada valor

```python
df['maritalstatus'].value_counts()
```

# Cambiar 'Never-Married' a 'Single'

```python
df['maritalstatus'] = df['maritalstatus'].replace('Never-Married', 'Single')
```

# Nueva Frecuencia

```python
df['maritalstatus'].value_counts()
```

# Ejercicio 2: Errores Tipográficos y Valores Inconsistentes: highestdegrees

# Nuevos valores

```python
df['highestdegree'].unique()
```

# Frecuencia

```python
df['highestdegree'].value_counts()
```

Limpieza de categorias

```python
df['highestdegree'] = df['highestdegree'].str.replace(r'^\d+\.\s*', '', regex=True)
df['highestdegree'] = df['highestdegree'].str.lower()
```

# Nuevos valores unicos

```python
df['highestdegree'].unique()
```

# Ejercicio 03: Alta Cardinalidad - Frecuencias Bajas

# Creacion de nueva columna marital_grouped

```python
df['marital_grouped'] = df['maritalstatus'].replace({
    'Widowed': 'Single/No Partner',
    'Single': 'Single/No Partner'
})
```

# Valores unicos

```python
df['marital_grouped'].unique()
```

# Frecuencia

```python
df['marital_grouped'].value_counts()
```

# Ejercicio 04: Tipos de Datos
