# Recomendación de cursos en plataforma e-learning

```python
import numpy as np
import pandas as pd

# =========================
# MATRIZ DE EJEMPLO
# =========================
# (Puedes cambiarla si te dieron una específica)

A = np.array([
    [5, 4, 4, 1, 0, 0, 0, 0, 2, 1, 0, 0, 3, 2, 4],
    [4, 5, 5, 2, 0, 0, 0, 1, 2, 2, 0, 0, 3, 3, 4],
    [0, 0, 1, 4, 5, 5, 4, 4, 5, 4, 3, 2, 0, 0, 2],
    [0, 0, 0, 5, 4, 4, 5, 5, 4, 5, 3, 3, 0, 0, 2],
    [3, 2, 3, 0, 0, 1, 1, 0, 0, 1, 4, 5, 5, 4, 3],
    [2, 3, 2, 0, 1, 0, 0, 1, 0, 0, 5, 4, 4, 5, 3]
])

print("Matriz original:")
print(A)

# =========================
# PARTE 2: SVD
# =========================
U, S, VT = np.linalg.svd(A)

print("\nDimensiones:")
print("U:", U.shape)
print("S:", S.shape)
print("VT:", VT.shape)

print("\nValores singulares:")
print(S)

# =========================
# PARTE 3: APROXIMACIÓN
# =========================
k = 2  # puedes cambiar a 3

S_k = np.zeros((k, k))
np.fill_diagonal(S_k, S[:k])

U_k = U[:, :k]
VT_k = VT[:k, :]

A_k = U_k @ S_k @ VT_k

print("\nMatriz aproximada:")
print(np.round(A_k, 2))

# =========================
# PARTE 4: ERROR
# =========================
def error_relativo(A, A_k):
    return np.linalg.norm(A - A_k) / np.linalg.norm(A)

print("\nErrores para diferentes k:")

for k in range(1, 6):
    S_k = np.zeros((k, k))
    np.fill_diagonal(S_k, S[:k])
    A_k = U[:, :k] @ S_k @ VT[:k, :]
    error = error_relativo(A, A_k)
    print(f"k={k} -> error={error:.4f}")

# =========================
# PARTE 5: VALOR FALTANTE
# =========================
A_missing = A.copy()
A_missing[0, 4] = 0  # ocultamos un valor

U, S, VT = np.linalg.svd(A_missing)

k = 2
S_k = np.zeros((k, k))
np.fill_diagonal(S_k, S[:k])

A_reconstruida = U[:, :k] @ S_k @ VT[:k, :]

print("\nValor estimado:")
print(A_reconstruida[0, 4])

# =========================
# PARTE 6: FACTORES LATENTES
# =========================
print("\nPrimer factor (U):")
print(U[:, 0])

print("\nPrimer factor (VT):")
print(VT[0])
```
