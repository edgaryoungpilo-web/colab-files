# Ejercicio datos academicos estudiantes

```python
#  Carga Librerias
import numpy as np
import pandas as pd
```

```python
# Vectores de pesos
pesos = np.array([0.12, 0.08, 0.17, 0.17, 0.10,
                  0.03, 0.04, 0.04, 0.05, 0.20])
```

```python
# Cargar archivo Excel
df = pd.read_excel('/content/drive/MyDrive/ClassFiles/Tablas de Datos Acad√©micos.xls')

# Extraer solo las calificaciones
Z = df.iloc[:, 2:].to_numpy()

Z
```

```python
# Pregunta 1
promedios_ponderados = pesos @ Z
```

```python
# Pregunta 2
promedios_sin_pesos = Z.mean(axis=0)
```

```python
# Pregunta 3
resultados = pd.DataFrame({
    "Promedio ponderado": promedios_ponderados,
    "Promedio sin pesos": promedios_sin_pesos
})

resultados

# Diferencia de promedios
resultados["Diferencia"] = (
    resultados["Promedio ponderado"] -
    resultados["Promedio sin pesos"]
)

resultados
```

```python
# Pregunta 4
correlacion = np.corrcoef(Z)
correlacion
```

```python
# Pregunta 5
from sklearn.decomposition import PCA

pca = PCA()
pca.fit(Z.T)

# Varianza
pca.explained_variance_ratio_

# Numero de componentes
np.cumsum(pca.explained_variance_ratio_)
```

```python
# Pregunta 6
#Estimar datos no almacenados
from sklearn.linear_model import LinearRegression

X = Z[:-1, :].T   # Todas menos una asignatura
y = Z[-1, :]      # Asignatura a estimar

modelo = LinearRegression()
modelo.fit(X, y)

y_estimado = modelo.predict(X)
```
