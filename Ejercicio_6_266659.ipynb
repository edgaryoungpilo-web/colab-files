# Manejo de librerias Panda

```python
# Direccion de la carpeta y asignacion de librerias

from google.colab import drive
drive.mount('/content/drive')

import pandas as pd
import numpy as np

# Ruta del archivo (ajusta según tu carpeta)
ruta = '/content/drive/MyDrive/ClassFiles/Titanic-Dataset.csv'
df = pd.read_csv(ruta)

df.head(890)
```

# Ejercicio 1 - Supervivencia por Sexo y Clase

```python
# Ejercicio 1 - Supervivencia por combinación de Sex y Pclass

supervivencia_sex_clase = df.groupby(['Sex', 'Pclass'])['Survived'].mean()
supervivencia_sex_clase

# Identificar la combinación con mayor y menor supervivencia
mayor_supervivencia = supervivencia_sex_clase.idxmax(), supervivencia_sex_clase.max()
menor_supervivencia = supervivencia_sex_clase.idxmin(), supervivencia_sex_clase.min()

print("Mayor tasa de supervivencia:", mayor_supervivencia)
print("Menor tasa de supervivencia:", menor_supervivencia)
```

# Ejercicio 2 - Identificacion de familias grandes

```python
# Creacion de columna
df['FamilySize'] = df['SibSp'] + df['Parch']

# Filtracion de familias grandes
familias_grandes = df[df['FamilySize'] > 3]

# Numero de pasajeros en familias grandes
num_familias_grandes = familias_grandes.shape[0]
print("Pasajeros en familias grandes:", num_familias_grandes)

# Proporcion de supervivencia en familias grandes
proporcion_supervivencia_familias_grandes = familias_grandes['Survived'].mean()
print("Proporción de supervivencia en familias grandes:", proporcion_supervivencia_familias_grandes)
```

 # Ejercicio 3 - Segmentacion por grupos de edad

```python
# Clasificacion de edad
def clasificar_edad(edad):
    if edad < 18:
        return 'Menor de edad'
    else:
        return 'Mayor de edad'
# Nueva Columna
df['GrupoEdad'] = df['Age'].apply(lambda x: clasificar_edad(x) if not np.isnan(x) else np.nan)
df[['Age', 'GrupoEdad']].head(10)
```

# Ejercicio 4 - Comparacion de promedios

```python
# Promedios Usados Numpy
promedio_age_numpy = np.nanmean(df['Age'])
promedio_fare_numpy = np.nanmean(df['Fare'])

# Promedios usando Pandas
promedio_age_pandas = df['Age'].mean()
promedio_fare_pandas = df['Fare'].mean()

# Coparacion
print("Edad - NumPy:", promedio_age_numpy)
print("Edad - Pandas:", promedio_age_pandas)

print("\nFare - NumPy:", promedio_fare_numpy)
print("Fare - Pandas:", promedio_fare_pandas)

```
