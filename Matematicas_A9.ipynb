# Matematicas A9 - Datos compramos el producto

# Importacion de librerias

```python
# Librerias

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Cargar Excel

```python
# Cargar Excel

df = pd.read_excel("/content/drive/MyDrive/Colab Notebooks/compramos_el_producto.xlsx")

print("Columnas del archivo:")
print(df.columns)
```

# Renombrar columnas

```python
# Columnas

df = df.rename(columns={
    'salario en miles': 'x',
    'probabilidad de compra': 'prob'
})
```

# Buscar valor mas cercano a 0.5

```python
# Probabilidad

# Buscar el valor más cercano a 0.5
indice = (df['prob'] - 0.5).abs().idxmin()
x0 = df.loc[indice, 'x']

print("x0 =", x0)
```

# Funcion logistica real

```python
# Funcion logistica

def f(x):
    return 1 / (1 + np.exp(-(x - x0)))
```

# Polinomio de Tylor

```python
# Polinomio

def P3(x):
    return 0.5 + 0.25*(x - x0) - (1/48)*(x - x0)**3
```

# Calcular aproximacion y error

```python
# Calculo de la aproximacion

df['Real'] = f(df['x'])
df['Aprox_Taylor'] = P3(df['x'])
df['Error_Absoluto'] = np.abs(df['Real'] - df['Aprox_Taylor'])

print("\nTabla final:")
display(df)
```

# Grafica

```python
# Grafica

plt.figure(figsize=(8,6))
plt.plot(df['x'], df['Real'], marker='o', label='Función Real')
plt.plot(df['x'], df['Aprox_Taylor'], marker='s', label='Taylor Orden 3')
plt.xlabel("Salario (en miles)")
plt.ylabel("Probabilidad")
plt.title("Función Logística vs Aproximación de Taylor")
plt.legend()
plt.grid()
plt.show()
```

# Error promedio

```python
print("\nError promedio:", df['Error_Absoluto'].mean())
```
