# Practica 11 - Seaborn y Subplots

# Librerias

```python
# Importar librerías
import seaborn as sns
import matplotlib.pyplot as plt

# Cargar el dataset
df = sns.load_dataset("penguins")
```

# Graficos

```python
# Crear figura con 2 filas y 2 columnas
fig, ax = plt.subplots(2, 2, figsize=(12, 10))

# Histograma de longitudes de aletas (0,0)
sns.histplot(data=df, x="flipper_length_mm", bins=20, kde=True,
             color="skyblue", ax=ax[0, 0])
ax[0, 0].set_title("Distribución de Longitud de Aletas")
ax[0, 0].set_xlabel("Longitud de aleta (mm)")
ax[0, 0].set_ylabel("Frecuencia")

# Boxplot de masa corporal por especie (0,1)
sns.boxplot(data=df, x="species", y="body_mass_g",
            palette="pastel", ax=ax[0, 1])
ax[0, 1].set_title("Masa Corporal por Especie")
ax[0, 1].set_xlabel("Especie")
ax[0, 1].set_ylabel("Masa corporal (g)")

# Scatterplot masa corporal vs longitud de aletas (1,0)
sns.scatterplot(data=df, x="flipper_length_mm", y="body_mass_g",
                hue="species", palette="deep", ax=ax[1, 0])
ax[1, 0].set_title("Masa Corporal vs Longitud de Aletas")
ax[1, 0].set_xlabel("Longitud de aleta (mm)")
ax[1, 0].set_ylabel("Masa corporal (g)")

# Violin plot de masa corporal por especie (1,1)
sns.violinplot(data=df, x="species", y="body_mass_g",
               palette="muted", ax=ax[1, 1])
ax[1, 1].set_title("Distribución de Masa Corporal por Especie")
ax[1, 1].set_xlabel("Especie")
ax[1, 1].set_ylabel("Masa corporal (g)")

# Ajustar diseño
plt.tight_layout()

# Mostrar figura
plt.show()
```
