# Ejercicio 1. Duplicidad Parcial Una columna: Contar y Filtrar los registros duplicados en Duration



Carga el conjunto de datos dirty.csv y verifica que su carga sea correcta.


```python
# Cargar dataset
ruta = "/content/drive/MyDrive/Colab Notebooks/dirtydata.csv"
df = pd.read_csv(ruta)

# Verificar que se cargó correctamente
print(df.head())
```

Cuenta la cantidad de registros duplicados utilizando solo la información de columna Duration e imprime


```python
# Contar registros duplicados solo en Duration
duplicados_duration = df.duplicated(subset=['Duration']).sum()
print("Cantidad de duplicados en Duration:", duplicados_duration)
```

Utiliza value_counts() e imprime la cantidad de registros que hay por cada valor único.


```python
# Cantidad de registros por cada valor único
print("\nConteo por valores únicos de Duration:")
print(df['Duration'].value_counts())
```

Filtra los datos y obten los registros sin duplicados


```python
# Filtrar registros sin duplicados
df_sin_duplicados_duration = df.drop_duplicates(subset=['Duration'])

print("\nDataset sin duplicados en Duration:")
print(df_sin_duplicados_duration)
```

# Ejercicio 2. Duplicidad parcial de dos columnas

Carga y verificacion de dataset

```python
# Cargar dataset
df = pd.read_csv(ruta)

# Verificar carga
print(df.head())
```

Cuenta la cantidad de registros duplicados utilizando la información de las columnas Pulse y Maxpulse

```python
# Contar duplicados usando Pulse y MaxPulse
duplicados_pulse = df.duplicated(subset=['Pulse','Maxpulse']).sum()
print("Cantidad de duplicados en Pulse y MaxPulse:", duplicados_pulse)
```

Utiliza value_counts() e imprime la cantidad de registros que hay por cada valor único.


```python
# Conteo de valores únicos
print("\nConteo por combinaciones Pulse y MaxPulse:")
print(df.value_counts(subset=['Pulse','Maxpulse']))
```

Filtra los datos y obten los registros cuando keep = 'last'


```python
# Filtrar duplicados quedándonos con el último
df_pulse_last = df.drop_duplicates(subset=['Pulse','Maxpulse'], keep='last')

print("\nDatos sin duplicados (conservando última ocurrencia):")
print(df_pulse_last)
```

# Ejercicio 3. Duplicidad Total o Exacta: Contar y Filtrar los registros duplicados



```python
# Cargar dataset
df = pd.read_csv(ruta)

# Verificar carga
print(df.head())
```

Cuenta la cantidad de registros duplicados exactos (todas las columnas)


```python
# Contar duplicados exactos
duplicados_totales = df.duplicated().sum()
print("Cantidad de duplicados exactos:", duplicados_totales)
```

Utiliza sum() e imprime la cantidad de registros con duplicidad exacta


```python
# Filtrar duplicados conservando última ocurrencia
df_sin_duplicados_totales = df.drop_duplicates(keep='last')

print("\nDataset sin duplicados exactos:")
print(df_sin_duplicados_totales)
```

# Ejercicio 4. Cálculo de la media con duplicados y sin duplicados



```python
# Cargar dataset
df = pd.read_csv(ruta)

# Verificar carga
print(df.head())
```

Calcula e imprime la media de las columnas Pulse y MaxPulse


```python
# Media con duplicados
media_con_dup = df[['Pulse','Maxpulse']].mean()
print("\nMedia con duplicados:")
print(media_con_dup)
```

Identifica los valores duplicados en las columnas Pulse y MaxPulse (Duplicidad Parcial) y elimina los registro repetidos. Quédate con las últimas ocurrencias


```python
# Eliminar duplicados en Pulse y MaxPulse (quedarse con la última ocurrencia)
df_sin_dup = df.drop_duplicates(subset=['Pulse','Maxpulse'], keep='last')
```

Utiliza el conjunto anterior sin datos duplicados (recuerda que solo quedaron las últimas ocurrencias de los datos) y calcula la media de las columnas Pulse y MaxPulse

```python
# Media sin duplicados
media_sin_dup = df_sin_dup[['Pulse','Maxpulse']].mean()
print("\nMedia sin duplicados:")
print(media_sin_dup)
```
