# Practica 18 Datos temporales

# Librerias

```python
# Importar librerías
import pandas as pd
```

# Ejercicio 1: Tipo de datos correcto

```python
# Cargar dataset
df = pd.read_csv('/content/drive/MyDrive/ClassFiles/.ipynb_checkpoints/Airdata.csv')

# Convertir la columna DateTime a tipo datetime si no lo está
df['DateTime'] = pd.to_datetime(df['DateTime'])

# Mostrar tipos de datos
print("Tipos de datos del DataFrame:")
print(df.dtypes)

# Imprimir primeras 3 filas
print("\nPrimeras 3 filas:")
print(df.head(3))
```

# Ejercicio 2: Extraer componentes de fecha y tiempo

```python
# Extraer componentes
df['Year'] = df['DateTime'].dt.year
df['Month'] = df['DateTime'].dt.month
df['Day'] = df['DateTime'].dt.day
df['Hour'] = df['DateTime'].dt.hour

# Mostrar primeras 3 filas
print("\nPrimeras 3 filas con componentes de fecha:")
print(df.head(3))
```

# Ejercicio 3: Aritmetica - suma de intervalos

```python
# Crear intervalo de tiempo
intervalo = pd.Timedelta(days=10, hours=7, minutes=15)

# Sumar intervalo a DateTime
df['DateTime_plus'] = df['DateTime'] + intervalo

# Mostrar columnas
print("\nDateTime original y DateTime con intervalo sumado:")
print(df[['DateTime','DateTime_plus']].head())
```

# Ejercicio 4: Timestamp

```python
# Convertir DateTime a timestamp Unix
df['Timestamp_Unix'] = df['DateTime'].astype('int64') // 10**9

# Mostrar primeras 3 filas
print("\nDateTime y Timestamp Unix:")
print(df[['DateTime','Timestamp_Unix']].head(3))
```

# Ejercicio 5: Filtrado de fechas

```python
# Filtrar fechas mayores a 2020-11-01
filtro = df[df['DateTime'] > '2020-11-01']

# Mostrar resultado
print("\nFechas mayores a 2020-11-01:")
print(filtro)
```
