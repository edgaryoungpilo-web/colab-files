# Seccion 1 - Importacion de librerias

```python
# Importacion
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.datasets import load_iris
```

# Seccion 2 - Carga del Database Iris

```python
# Carga de Database
iris = load_iris(as_frame=True)
df = iris.frame

# Agregar columna con el nombre de la especie
df['species'] = df['target'].map(dict(enumerate(iris.target_names)))

# Eliminar la columna target (ya no es necesaria)
df = df.drop(columns='target')
```

# Seccion 3 - Exploracion del Database

```python
# 3. Exploración del dataset

# Mostrar las primeras filas
print("Primeras filas del dataset:")
print(df.head())

# Información general del dataset
print("\nInformación general del dataset:")
df.info()
```

# Seccion 4 - Graficos con Matplolib

```python
# 4. Gráficos con Matplotlib
# a) Gráfico de barras
# Promedio de largo y ancho del pétalo por especie

means = df.groupby('species')[['petal length (cm)', 'petal width (cm)']].mean()

plt.figure()
means.plot(kind='bar')
plt.title("Promedio de largo y ancho del pétalo por especie")
plt.xlabel("Especie")
plt.ylabel("Promedio (cm)")
plt.legend(title="Medida")
plt.show()
```

```python
# b) Histograma
# Distribución del largo del sépalo
# --------------------------------
plt.figure()
plt.hist(df['sepal length (cm)'], bins=20)
plt.title("Distribución del largo del sépalo")
plt.xlabel("Sepal length (cm)")
plt.ylabel("Frecuencia")
plt.show()
```

```python
# c) Gráfico de dispersión
# Relación entre largo y ancho del pétalo
# --------------------------------
plt.figure()
for species in df['species'].unique():
    subset = df[df['species'] == species]
    plt.scatter(
        subset['petal length (cm)'],
        subset['petal width (cm)'],
        label=species
    )

plt.title("Relación entre largo y ancho del pétalo")
plt.xlabel("Petal length (cm)")
plt.ylabel("Petal width (cm)")
plt.legend(title="Especie")
plt.show()
```

```python
# d) Gráfico de cajas (box plot)
# Distribución del sépalo por especie
plt.figure()
df.boxplot(
    column=['sepal length (cm)', 'sepal width (cm)'],
    by='species'
)
plt.title("Distribución de medidas del sépalo por especie")
plt.suptitle("")  # Quita el título automático
plt.xlabel("Especie")
plt.ylabel("Medida (cm)")
plt.show()
```

```python
# e) Gráfico de líneas
# Tendencia de la longitud del pétalo
plt.figure()
plt.plot(df.index, df['petal length (cm)'])
plt.title("Tendencia de la longitud del pétalo")
plt.xlabel("Índice de observación")
plt.ylabel("Petal length (cm)")
plt.show()
```
