# Uso de mascaras booleanas

```python
import numpy as np

# 1. Crear un arreglo unidimensional con 15 números aleatorios entre 1 y 50
arreglo = np.random.randint(1, 51, size=15)

print("Arreglo original:")
print(arreglo)

# 2. Crear la máscara booleana (valores mayores que 25)
mascboo = arreglo > 25

print("\nMáscara booleana (mayores que 25):")
print(mascboo)

# 3. Seleccionar los elementos usando la máscara
elementos_mayores_25 = arreglo[mascboo]

print("\nElementos mayores que 25:")
print(elementos_mayores_25)

# 4. Contar cuántos elementos son mayores que 25
cantidad = np.sum(mascboo)

print("\nCantidad de elementos mayores que 25:", cantidad)
```
