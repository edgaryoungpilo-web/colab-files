# Practica 4 - Manipulacion de matrices

```python
import numpy as np

# 1. Crear dos matrices 3x3 con valores aleatorios entre 1 y 10
matriz1 = np.random.randint(1, 11, size=(3, 3))
matriz2 = np.random.randint(1, 11, size=(3, 3))

print("Matriz 1:")
print(matriz1)

print("\nMatriz 2:")
print(matriz2)

# 2. Suma de matrices
suma = matriz1 + matriz2
print("\nSuma de las matrices:")
print(suma)

# 3. Resta de matrices (matriz1 - matriz2)
resta = matriz1 - matriz2
print("\nResta de las matrices (Matriz1 - Matriz2):")
print(resta)

# 4. Multiplicar la primera matriz por 2
multiplicacion_escalar = matriz1 * 2
print("\nMatriz 1 multiplicada por 2:")
print(multiplicacion_escalar)

# 5. Multiplicación matricial
multiplicacion_matricial = np.dot(matriz1, matriz2)
print("\nMultiplicación matricial (Matriz1 x Matriz2):")
print(multiplicacion_matricial)
```
